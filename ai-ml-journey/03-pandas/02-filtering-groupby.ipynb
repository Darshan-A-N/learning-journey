```python
import pandas as pd
#create smaple data
data={
    "Name":["Alice","Bob","Charlie","David","Eve","Frank"],
    "Branch":["ECE","CSE","ECE","AI-ML","AI-ML","AI-ML"],
    "Year":[2,3,2,2,2,3],
    "CGPA":[8.5,9.2,7.8,8.8,8.0,9.5]
}
df=pd.DataFrame(data)
print(df)
```

1.Filtering Data

```python
#ECE students only
aiml_students = df[df["Branch"]=="AI-ML"]
print("AI-ML Students are there:\n",aiml_students)
```

```python
#CGPA > 8.5
top_students = df[df["CGPA"]>=8.5]
print("Top STudents Scored More than 8.5 \n",top_students)
```

```python
#Multiple Conditions: AI-ML + Year2
aiml_year2=df[(df["Year"]==2) &(df["Branch"]=="AI-ML")]
print("AI-ml & in Sec Year Students\n",aiml_year2)
```

2. GroupBy Operations

```python
#Average CGPA by Branch
branch_avg = df.groupby("Branch")["CGPA"].mean().reset_index(name="Average CGPA")
print("Branch Wise Average  \n",branch_avg)
```

```python
#multiple aggreations
branch_stats=df.groupby("Branch")["CGPA"].agg(["mean","max","min","count"])
print("multiple aggreation\n",branch_stats)
```

4. Challenge: Department Analysis

```python
employees = pd.DataFrame({
    "Name": ["Raj", "Priya", "Amit", "Sneha", "Vikram", "Anita", "Ravi"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 65000, 58000],
    "Experience": [3, 2, 5, 4, 3, 6, 4]
})

# 1. Average salary by department
print("Average by dept:\n", employees.groupby("Department")["Salary"].mean())

# 2. Highest paid in each department
print("\nHighest paid:\n", employees.loc[employees.groupby("Department")["Salary"].idxmax()])

# 3. Salary > 55000 AND Experience > 3
print("\nSenior high earners:\n", employees[(employees["Salary"] > 55000) & (employees["Experience"] > 3)])

# 4. Bonus column
employees["Bonus"] = employees.apply(lambda x: x["Salary"] * 0.10 if x["Experience"] > 4 else x["Salary"] * 0.05, axis=1)
print("\nWith Bonus:\n", employees)

# 5. Sort by dept then salary desc
print("\nSorted:\n", employees.sort_values(["Department", "Salary"], ascending=[True, False]))

# 6. Count per dept
print("\nCount:\n", employees["Department"].value_counts())
```
