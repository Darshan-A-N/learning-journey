# Day 17: Linear Regression

## The Idea
Find the best straight line through data points.

Equation: **y = mx + b**
- y = predicted value
- m = slope (how steep)
- b = intercept (where line crosses y-axis)
- x = input feature

```python
import numpy as np
import matplotlib.pyplot as plt
# Simple data: Study hours vs Exam score
X = np.array([1, 2, 3, 4, 5, 6, 7, 8])  # Hours studied
y = np.array([50, 55, 60, 65, 70, 75, 80, 85])  # Exam score
plt.scatter(X, y, color='blue', s=100)
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Study Hours vs Exam Score")
plt.show()
```

## Finding the Line: Manual Calculation

Formula for slope (m) and intercept (b):
- m = Σ((x - x_mean)(y - y_mean)) / Σ((x - x_mean)²)
- b = y_mean - m * x_mean

```python
# Calculate manually
x_mean = np.mean(X)
y_mean = np.mean(y)
# Calculate slope (m)
numerator = np.sum((X - x_mean) * (y - y_mean))
denominator = np.sum((X - x_mean) ** 2)
m = numerator / denominator
# Calculate intercept (b)
b = y_mean - m * x_mean
print(f"Slope (m): {m:.2f}")
print(f"Intercept (b): {b:.2f}")
print(f"Equation: y = {m:.2f}x + {b:.2f}")
# Predict
def predict(x):
    return m * x + b
# Test: Predict score for 5.5 hours
print(f"\nPredicted score for 5.5 hours: {predict(5.5):.2f}")
```

```python
# Plot the line
y_pred = predict(X)
plt.scatter(X, y, color='blue', s=100, label='Actual')
plt.plot(X, y_pred, color='red', linewidth=2, label='Predicted Line')
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Linear Regression: Manual")
plt.legend()
plt.show()
```

## Using Scikit-Learn (The Easy Way)

```python
from sklearn.linear_model import LinearRegression
# Reshape X (sklearn needs 2D array)
X_reshaped = X.reshape(-1, 1)
# Create and train model
model = LinearRegression()
model.fit(X_reshaped, y)
# Get parameters
print(f"Sklearn Slope: {model.coef_[0]:.2f}")
print(f"Sklearn Intercept: {model.intercept_:.2f}")
# Predict
prediction = model.predict([[5.5]])
print(f"Predicted score for 5.5 hours: {prediction[0]:.2f}")
# Plot
y_sklearn = model.predict(X_reshaped)
plt.scatter(X, y, color='blue', s=100, label='Actual')
plt.plot(X, y_sklearn, color='green', linewidth=2, label='Sklearn Line')
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Linear Regression: Scikit-Learn")
plt.legend()
plt.show()
```
