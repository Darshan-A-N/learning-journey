#  Model Evaluation Metrics

How good is our model? We need numbers to measure it.

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# Actual vs Predicted
y_actual = np.array([50, 55, 60, 65, 70, 75, 80, 85])
y_predicted = np.array([52, 54, 61, 64, 72, 74, 81, 84])
```

## 1. MAE — Mean Absolute Error

Average of absolute differences.
- Easy to understand
- Same unit as target
- Robust to outliers

Formula: mean(|actual - predicted|)

```python
# Manual calculation
mae_manual = np.mean(np.abs(y_actual - y_predicted))
print(f"MAE (manual): {mae_manual:.2f}")
# Using sklearn
mae_sklearn = mean_absolute_error(y_actual, y_predicted)
print(f"MAE (sklearn): {mae_sklearn:.2f}")
```

## 2. MSE — Mean Squared Error

Average of squared differences.
- Punishes large errors more
- Harder to interpret (squared units)

Formula: mean((actual - predicted)²)

```python
# Manual
mse_manual = np.mean((y_actual - y_predicted) ** 2)
print(f"MSE (manual): {mse_manual:.2f}")

# Sklearn
mse_sklearn = mean_squared_error(y_actual, y_predicted)
print(f"MSE (sklearn): {mse_sklearn:.2f}")
```

## 3. RMSE — Root Mean Squared Error

Square root of MSE.
- Same unit as target
- Punishes large errors

Formula: sqrt(MSE)

```python
rmse_manual = np.sqrt(mse_manual)
print(f"RMSE (manual): {rmse_manual:.2f}")

rmse_sklearn = np.sqrt(mean_squared_error(y_actual, y_predicted))
print(f"RMSE (sklearn): {rmse_sklearn:.2f}")
```

## 4. R² Score — Coefficient of Determination

How well model explains variance.
- 1.0 = perfect prediction
- 0.0 = model is as good as predicting the mean
- Negative = worse than predicting mean

Formula: 1 - (MSE of model / MSE of predicting mean)

```python
# Manual
y_mean = np.mean(y_actual)
ss_total = np.sum((y_actual - y_mean) ** 2)
ss_residual = np.sum((y_actual - y_predicted) ** 2)
r2_manual = 1 - (ss_residual / ss_total)
print(f"R² (manual): {r2_manual:.4f}")

# Sklearn
r2_sklearn = r2_score(y_actual, y_predicted)
print(f"R² (sklearn): {r2_sklearn:.4f}")
```

## 5. Compare Two Models

Which model is better? Lower MAE/MSE/RMSE, higher R².

```python
# Model A predictions
y_pred_a = np.array([52, 54, 61, 64, 72, 74, 81, 84])

# Model B predictions (worse)
y_pred_b = np.array([48, 50, 58, 62, 68, 70, 78, 82])

print("Model A:")
print(f"  MAE: {mean_absolute_error(y_actual, y_pred_a):.2f}")
print(f"  RMSE: {np.sqrt(mean_squared_error(y_actual, y_pred_a)):.2f}")
print(f"  R²: {r2_score(y_actual, y_pred_a):.4f}")

print("\nModel B:")
print(f"  MAE: {mean_absolute_error(y_actual, y_pred_b):.2f}")
print(f"  RMSE: {np.sqrt(mean_squared_error(y_actual, y_pred_b)):.2f}")
print(f"  R²: {r2_score(y_actual, y_pred_b):.4f}")

print("\nWinner: Model A (lower error, higher R²)")
```
