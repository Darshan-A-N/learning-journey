# Day 16: What is Machine Learning?

## Definition
Machine Learning = Computer learns from data without being explicitly programmed

## Analogy
Traditional Programming:
Input → Rules (coded by human) → Output

Machine Learning:
Input + Output → Model learns Rules → New Input → Predicted Output

```python
# Traditional Programming Example
def calculate_area(length, width):
    return length * width  # Rule is explicitly coded

# ML Example: We give data, model finds the rule
# Data: [(2,3,6), (4,5,20), (3,4,12)] → Model learns: area = length * width
```

## Types of Machine Learning

| Type | What | Example |
|------|------|---------|
| **Supervised** | Learn from labeled data (input + correct output) | Predict house price from size |
| **Unsupervised** | Find patterns in unlabeled data | Group customers by behavior |
| **Reinforcement** | Learn by trial and error | Game playing AI |

We'll focus on **Supervised Learning** first.

## Supervised Learning: Two Types

### 1. Regression → Predict continuous number
- House price, temperature, salary
- Output: Any number (50,000 or 8.5 or 25.3)

### 2. Classification → Predict category
- Spam or not spam, cat or dog, pass or fail
- Output: Specific category (0 or 1, "yes" or "no")

```python
# Regression Example: Predict exam score from study hours
# Data: (hours_studied, exam_score)
# (2, 55), (3, 60), (4, 65), (5, 72), (6, 78), (7, 85), (8, 90)
# New input: 5.5 hours → Predict score?

# Classification Example: Pass or fail based on attendance
# Data: (attendance%, result)
# (90, "Pass"), (80, "Pass"), (60, "Fail"), (95, "Pass"), (50, "Fail")
# New input: 75% → Predict: Pass or Fail?
```

## The ML Workflow

1. **Get Data** → Collect or load dataset
2. **Clean Data** → Handle missing values, outliers (you already know this!)
3. **Split Data** → Training set (80%) + Testing set (20%)
4. **Choose Model** → Linear Regression, Decision Tree, etc.
5. **Train Model** → Feed training data, model learns patterns
6. **Evaluate** → Test on unseen data, check accuracy
7. **Predict** → Use model on new data
