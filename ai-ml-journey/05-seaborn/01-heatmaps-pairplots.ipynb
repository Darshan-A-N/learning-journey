 Day 8: Seaborn - Heatmaps, Pair Plots & Distribution

Learning advanced visualization with Seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#set style
sns.set_style("whitegrid")
%matplotlib inline
```

1. Heatmap - Correlation Matrix

```python
#create sample student data
data={
    "Math":[88,89,89,87,57,98,95,94],
    "Physics": [90, 88, 85, 92, 80, 85, 88, 95],
    "Chemistry": [78, 95, 80, 90, 85, 88, 92, 85],
    "Python": [92, 85, 88, 95, 78, 90, 85, 92]
}
df=pd.DataFrame(data)
#calculate coreleation
corr=df.corr()
print("Corelation Matrix:\n",corr)
#Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm',linewidths=0.5,fmt='.2f')
plt.title("subject correlation Heatmap")
plt.show()
```

 2. Pair Plot - Relationships Between Variables

```python
#add a " Total " column
df["Total"]=df.sum(axis=1)
#pair plot
sns.pairplot(df,diag_kind="kde",plot_kws={"alpha":0.6})
plt.suptitle("pair plot of subjects",y=1.02)
plt.show()
```

 3. Distribution Plots

```python
# Histogram with KDE
plt.figure(figsize=(10, 5))
sns.histplot(df["Math"], kde=True, bins=8, color="blue")
plt.title("Math Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Count")
plt.show()
```

```python
# Box plot
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[["Math", "Physics", "Chemistry", "Python"]])
plt.title("Subject Marks - Box Plot")
plt.ylabel("Marks")
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(8, 5))
sns.violinplot(data=df[["Math", "Physics", "Chemistry", "Python"]])
plt.title("Subject Marks - Violin Plot")
plt.ylabel("Marks")
plt.show()
```

4. Categorical Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Create DataFrame
students = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Henry"],
    "Branch": ["ECE", "CSE", "ECE", "MECH", "CSE", "ECE", "MECH", "CSE"],
    "CGPA": [8.5, 9.2, 7.8, 8.8, 8.0, 9.5, 8.2, 8.9]
})
# Bar plot
plt.figure(figsize=(8, 5))
sns.barplot(
    x="Branch",
    y="CGPA",
    hue="Branch",
    data=students,
    palette="viridis",
    legend=False
)
plt.title("Average CGPA by Branch")
plt.show()
# Strip plot
plt.figure(figsize=(8, 5))
sns.stripplot(
    x="Branch",
    y="CGPA",
    hue="Branch",
    data=students,
    palette="Set2",
    size=10,
    legend=False
)
plt.title("CGPA Distribution by Branch")
plt.show()
```

 5. Challenge: Complete EDA with Seaborn

```python
import numpy as np
# Load iris dataset (built into Seaborn)
iris = sns.load_dataset("iris")
print(iris.head())
print(iris.describe())
# 1. Correlation heatmap of numeric columns
numeric_cols = iris.select_dtypes(include=[np.number])
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap="RdYlBu", fmt=".2f")
plt.title("Iris Dataset Correlation")
plt.show()
# 2. Pair plot colored by species
sns.pairplot(iris, hue="species", palette="Set1", diag_kind="kde")
plt.suptitle("Iris Pair Plot by Species", y=1.02)
plt.show()
# 3. Distribution of sepal length by species
plt.figure(figsize=(10, 6))
sns.histplot(data=iris, x="sepal_length", hue="species", kde=True, bins=15)
plt.title("Sepal Length Distribution by Species")
plt.show()
# 4. Box plot of all features by species
plt.figure(figsize=(12, 8))
iris_melted = iris.melt(id_vars="species", var_name="Feature", value_name="Value")
sns.boxplot(x="Feature", y="Value", hue="species", data=iris_melted)
plt.title("Feature Comparison by Species")
plt.xticks(rotation=45)
plt.show()
```
