 Day 7: Matplotlib - Line, Bar, and Scatter Plots
Learning data visualization with Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

 1. Line Plot

```python
#simple line plot
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
#Styled Line Plot
days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
temperature=[28,30,32,29,31,33,30]
plt.plot(days,temperature,color='red',marker='o',linestyle='--',linewidth=2,markersize=8)
plt.title("Weakly Temperature")
plt.xlabel("Day")
plt.ylabel("Temperature (oc)")
plt.grid(True)
plt.show()
```

2. Bar Plot

```python
subjects=["Maths","Physics","Chemistry","Python"]
marks=[85,67,89,87]
plt.plot(subjects,marks,color='red',marker='o',linestyle='--',linewidth=2,markersize=8)
plt.bar(subjects,marks,color=['blue','green','orange','purple'])
plt.title("Subject Marks")
plt.xlabel("subject")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python
#horizontal bar
plt.barh(subjects,marks,color='teal')
plt.title("Subject marks - Horizontal")
plt.xlabel("Marks")
plt.show()
```

3. Scatter Plot

```python
#student data
study_hour=[2,3,4,5,6,7,8,9]
exam_score=[65,78,67,98,58,46,87,87]
plt.scatter(study_hour,exam_score,color='red',s=100,alpha=0.7)
plt.plot(study_hour,exam_score,color='red',marker='o',linestyle='--',linewidth=2,markersize=8)
plt.title("Study_hours vs Exam_Scores")
plt.xlabel("study Hours")
plt.ylabel("Exam_Score")
plt.grid(True)
plt.show()
```

4. Multiple Plots in One Figure

```python
fig, axes=plt.subplots(1,3,figsize=(14,4))
#liner plot
axes[0].plot(days,temperature,'o-',color='red')
axes[0].set_title("Temperature")
axes[0].set_xlabel("Day")
#bar plot
axes[1].bar(subjects,marks,color='blue')
axes[1].set_title("Marks")
axes[1].set_ylabel("score")
#Scatter plot
axes[2].scatter(study_hour,exam_score,color='green')
axes[2].set_title("study vs score")
axes[2].set_xlabel("Hours")
plt.tight_layout()
plt.show()
```

5. Challenge: Sales Data Visualization

```python
import pandas as pd

# Sales data
sales_data = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Revenue": [12000, 15000, 13000, 17000, 16000, 19000],
    "Expenses": [8000, 9000, 8500, 10000, 9500, 11000],
    "Customers": [150, 180, 160, 200, 190, 220]
})

# 1. Line plot: Revenue and Expenses over months
plt.figure(figsize=(10, 5))
plt.plot(sales_data["Month"], sales_data["Revenue"], marker='o', label='Revenue', color='green')
plt.plot(sales_data["Month"], sales_data["Expenses"], marker='s', label='Expenses', color='red')
plt.title("Revenue vs Expenses")
plt.xlabel("Month")
plt.ylabel("Amount ($)")
plt.legend()
plt.grid(True)
plt.show()

# 2. Bar plot: Customers per month
plt.figure(figsize=(8, 5))
plt.bar(sales_data["Month"], sales_data["Customers"], color='skyblue')
plt.title("Customers per Month")
plt.xlabel("Month")
plt.ylabel("Number of Customers")
plt.show()

# 3. Scatter plot: Revenue vs Customers
plt.figure(figsize=(8, 5))
plt.scatter(sales_data["Customers"], sales_data["Revenue"], s=100, color='purple')
plt.title("Revenue vs Customers")
plt.xlabel("Customers")
plt.ylabel("Revenue ($)")
plt.show()

# 4. Calculate profit and add bar plot
sales_data["Profit"] = sales_data["Revenue"] - sales_data["Expenses"]
plt.figure(figsize=(8, 5))
colors = ['green' if p > 0 else 'red' for p in sales_data["Profit"]]
plt.bar(sales_data["Month"], sales_data["Profit"], color=colors)
plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit ($)")
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
plt.show()
```
